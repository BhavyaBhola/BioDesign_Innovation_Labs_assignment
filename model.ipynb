```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
```

```python
df1 = pd.read_csv('reduced_df1.csv')
df2 = pd.read_csv('reduced_df2.csv')
df3 = pd.read_csv('reduced_df3.csv')
```

```python
print(df1.shape)
print(df2.shape)
print(df3.shape)
```

```python
Xtrain_df1, Xtest_df1, ytrain_df1, ytest_df1 = train_test_split(df1.drop('Colon_Cancer_Type', axis=1), df1['Colon_Cancer_Type'], test_size=0.2, random_state=42)
Xtrain_df2, Xtest_df2, ytrain_df2, ytest_df2 = train_test_split(df2.drop('Colon_Cancer_Type', axis=1), df2['Colon_Cancer_Type'], test_size=0.2, random_state=42)
Xtrain_df3, Xtest_df3, ytrain_df3, ytest_df3 = train_test_split(df3.drop('Colon_Cancer_Type', axis=1), df3['Colon_Cancer_Type'], test_size=0.2, random_state=42)
```

```python
rf1 = RandomForestClassifier(random_state=42)
rf2 = RandomForestClassifier(random_state=43)
rf3 = RandomForestClassifier(random_state=44)

rf1.fit(Xtrain_df1, ytrain_df1)
rf2.fit(Xtrain_df2, ytrain_df2)
rf3.fit(Xtrain_df3, ytrain_df3)
```

```python
Xtest_df1.to_numpy()
```

```python
pred1 = rf1.predict_proba(Xtest_df1[:1].to_numpy())
pred2 = rf2.predict_proba(Xtest_df2[:1].to_numpy())
pred3 = rf3.predict_proba(Xtest_df3[:1].to_numpy())
```

```python
final_probs =  [(pred1[:,0] + pred2[:,0] + pred3[:,0])/3, (pred1[:,1] + pred2[:,1] + pred3[:,1])/3, (pred1[:,2] + pred2[:,2] + pred3[:,2])/3]


print(final_probs)
```

```python
class_type = {
    0:'Type A',
    1:'Type B',
    2:'Type C'
}


print(f'The type of colon cancer is {class_type[np.argmax(final_probs)]}')
```
